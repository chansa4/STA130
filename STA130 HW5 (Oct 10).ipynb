**1) Understanding hypothesis testing** <br> 
**What is the key factor that makes the difference between ideas that can, and cannot be examined and tested statistically?** <br> 
For an idea to be testable, it must be able to be converted into measurable quantities, only this would allow a formation of a testable null hypothesis. <br> 
**What would you describe is the key "criteria" defining what a good null hypothesis is?** <br> 
First the null hypothsis must be able to be testable, meaning that it must have the abillity to collect and analyze relevant data. Secondly, it should be specific and clearly defines the population parameter and its hypothesized value. Lastly is must be able to gather evidence that contridicts the null hypothesis, as our goal is to reject the null. <br> 
**What is the difference between a null hypothesis and an alternative hypothesis in the context of hypothesis testing?** <br> 
A null hypothesis typically proposes that there is 'no effect,' 'no difference,' or 'no relationship' between two groups. It is a statement about a population parameter, and we aim to determine if the data we collect is strong enough to reject the null. An alternative hypothesis proposes that there is an effect, difference, or relationship, as it is the opposite of the null hypothesis. The alternative hypothesis is what we suspect or hope to be true; it is what we will consider if we find enough evidence to reject the null.

**2) "It is important to note that outcomes of tests refer to the population parameter, rather than the sample statistic! As such, the result that we get is for the population." - Explaination** <br>

$\mu$: The true mean of the entire population, this is often unknown and what we are trying to find <br> 
$\mu_0$: This is a hypothesized value of the population mean ($\mu$), this is what is assumed before doing the statistical test. <br>

The sentance "It is important to note that outcomes of tests refer to the population parameter, rather than the sample statistic! As such, the result that we get is for the population." This means that even though we are using sample datapoints ($x_i\!$) to perform these statistical hypothesis test and calculate sample mean($\bar x$), the main purpose of the test is to find out about the population mean ($\mu$), and not just the sample. These tests are trying to help us understand if the population mean ($\mu$), is equal to or different from the hypothesized value ($\mu_0$), based on evidence from the sample data. While we colelct and analyze sample data, the test result is to try and help us infer something about the larger population mean ($\mu$). While the tests conducted are based on the sample, the inference or conclusion we draw fromt he tests applies to the entire population. 

**4) Understanding p-values and null hypothesis**
A p-value is the probability of obtaining a test statistic at least as extreme as the one we actually observed (the one that we calculated using actual data) if we were to repeatedly sample from a population where the null hypothesis is true. The null hypothesis represents a default position in which there is no effect or difference. If we calculate a very small p-value, it could mean that the observed statistic we calculated is a very rare and unlikely event that we have encountered or that the null hypothesis is not true. The smaller the p-value, the less likely it is that we would be the sample to see such a rare result just by chance; therefore, we have more reason to reject it. Normally, we have a threshold (our significance level) such that if our p-value is equal to or less than this number, there is enough evidence to reject the null hypothesis.

**8) Fisher's Tea Experiment**<br> 
**Problem introduction** <br> 
Ronald Fisher and Dr. Muriel Bristol often enjoyed a cup of tea together. One day, Bristol claimed that she could taste the difference between tea poured with milk first and tea poured with the tea first. Skeptical, Fisher proposed a hypothesis test to examine the situation. He prepared 8 cups of tea, 4 with milk poured first and 4 with tea poured first, and Bristol correctly identified the pouring order for all 8 cups. Still uncertain, Fisher designed a statistical hypothesis test to determine whether her success could be attributed to chance. <br> 
<br> 
In our experiment, we replicate this idea with a larger sample of 80 STA130 students. Each student tastes one cup of tea and is asked to identify whether the milk or tea was poured first. Of these 80 students, 49 correctly identified the order. Similar to Fisher's original test, our goal is to perform a statistical analysis to determine whether the students are merely guessing, or if their correct responses suggest a genuine ability to detect the order of pouring. This experiment scales up Fisher's original study and applies the same statistical principles to evaluate whether the students performance is better than random guessing.
<br> 
$H_0$: The proportion of STA130 students who can corretctly identify the order of pouring milk or tea is equal to 0.5 <br>
$H_0$: $P = 0.5$ <br> 
<br> 
**Informal Interpretation**<br> 
The null hypothesis states that the students are just geussing when they identify the order in which the milk or tea was poured, meaning there is no real abillity to distiguish between the two methods. <br> 
<br> 
$H_A$: The proportion of STA130 students who can correctly identify the order of pouring milk or tea is greater than 0.5 <br> 
$H_A$: $P > 0.5$ <br> 

**Hypothesis testing using simulation** <br> 
Down below, we begin out hypothesis testing through simulation under the null hypothesis, where $p = 0.5$. We simulated 10000 samples, and gathered the p value. This is illustrated using a histogram generated from this above, with the red line being out observed proportion of $49/80$. We have decided to set the significance level at 0.05. 

```python
import numpy as np
import plotly.graph_objects as go

# Parameters
n = 80  # sample size
x = 49  # number of correct guesses
observed_proportion = x / n
num_simulations = 10000  # number of simulations

# Simulate trials under the null hypothesis
np.random.seed(42)  # for reproducibility
simulated_proportions = np.random.binomial(n, 0.5, num_simulations) / n

# Calculate the p-value
p_value = np.mean(simulated_proportions >= observed_proportion)

# Create a histogram using Plotly
fig = go.Figure()

# Add histogram for simulated proportions
fig.add_trace(go.Histogram(
    x=simulated_proportions,
    nbinsx=30,
    name='Simulated Proportions',
    marker_color='blue',
    opacity=0.75,
    histnorm='probability'  # Normalize the histogram to show probabilities
))

# Add line for observed proportion
fig.add_trace(go.Scatter(
    x=[observed_proportion, observed_proportion],
    y=[0, 0.2],  # Adjust y range as needed
    mode='lines',
    line=dict(color='red', width=2, dash='dash'),
    name='Observed Proportion'
))

# Update layout
fig.update_layout(
    title='Simulated Proportions of Correct Responses',
    xaxis_title='Proportion Correct',
    yaxis_title='Probability',
    barmode='overlay',
)

# Show the plot
fig.show(renderer ='png')

# Output results
print(f"Observed Proportion: {observed_proportion:.4f}")
print(f"P-value: {p_value:.4f}")
```

**Explaination of Code** <br> 
The parameters for the sample size (n), the number of correct guesses (x), and the number of simulations to be run (num_simulations) are set. The observed proportion of correct guesses is calculated by dividing x by n. Using NumPy, the code generates 10,000 simulated trials under the null hypothesis, which assumes that the true proportion of correct guesses is 0.5. Each simulation draws from a binomial distribution, where the number of successes is determined by n and the probability of success is 0.5. The simulated proportions of correct guesses are then calculated. Finally, the code computes the p-value by determining the proportion of simulated trials that yield a result greater than or equal to the observed proportion, indicating how likely the observed result would be if the null hypothesis were true.
