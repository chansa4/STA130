**4) R-squared vs. P-values**

**R-Squared**: With an R-squared value of 0.176, the model explains 17.6% of the variation in the dependent variable. This leaves 82.4% unexplained, as the model does not account for it.

**P-Values**: In this model, many coefficients are larger than 10 and have very small p-values, indicating they are statistically significant.

While these two statements may seem contradictory, they actually address different aspects of the model.

- **R-Squared**: The value of 0.176 tells us that the model as a whole only explains 17.6% of the variability in the dependent variable. So, while the model may be statistically significant, it doesn’t capture much of the outcome’s variability, suggesting that a large amount of the variability in the dependent variable remains unexplained.

- **P-Values**: The small p-values and large coefficients show that specific predictors have statistically significant relationships with the outcome. This means they likely influence the outcome variable in a meaningful way, even though the overall model fit (R-squared) is low.

The difference between the two statements is that R-squared measures how well the entire model explains the outcome, while p-values and coefficients indicate how much specific predictors matter in influencing the outcome, regardless of the model’s overall performance.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
# fail https://github.com/KeithGalli/pandas/blob/master/pokemon_data.csv
pokeaman = pd.read_csv(url) 
pokeaman
```

**7) Development of models** <br>
**Development of model5_linear_form from model3_fit and model4_fit**:  
Model 3 focuses only on Attack and Defense to predict HP; however, this doesn't fully capture the complexity of the data. The hypothesis testing shows that these variables have a significant impact, but the in-sample R-squared value (0.1477) is relatively low, suggesting that the model doesn't capture much of the variance in HP. This indicates that the model is likely underfitting the data by not accounting for enough complexity in the relationship. Model 4 attempts to improve this by adding more predictors and interaction terms to capture more relationships between these variables and HP. As seen, the in-sample R-squared (0.4671) has improved compared to Model 3. The increase in the in-sample R-squared suggests that the model has gained more predictive power. However, the out-of-sample R-squared (0.0025) dropped dramatically compared to Model 3, suggesting severe overfitting. The model performs very well on the training data but fails to generalize to the test data. In Model 5, some of the interactions from Model 4 were removed to avoid overfitting, but it still includes more predictors than Model 3. The in-sample R-squared (0.3920) and out-of-sample R-squared (0.3002) values show that the model is not as overfitted, and the out-of-sample performance is significantly better than Model 4's. This demonstrates improved generalizability. Model 5 represents a more balanced approach.

```python
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
```

```python
model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
yhat_model4 = model4_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model4_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model4)[0,1]**2)
```

```python
# Here's something a little more reasonable...
model5_linear_form = 'HP ~ Attack + Defense + Speed + Legendary'
model5_linear_form += ' + Q("Sp. Def") + Q("Sp. Atk")'
model5_linear_form += ' + C(Generation) + C(Q("Type 1")) + C(Q("Type 2"))'

model5_spec = smf.ols(formula=model5_linear_form, data=pokeaman_train)
model5_fit = model5_spec.fit()
model5_fit.summary()
yhat_model5 = model5_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model5_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model5)[0,1]**2)
```

**Development of model6_linear_form from model5_linear_form**: Model 6 builds on Model 5 by focusing on a more targeted selection of predictors (Attack, Speed, Special Defense, and Special Attack), which were key variables identified. Additionally, Model 6 introduces significant indicator variables that were highlighted in the previous models, such as Type 1 indicators and Generations. The in-sample R-squared for Model 6 (0.3326) shows a slight improvement compared to Model 5, and the out-of-sample R-squared for Model 6 (0.2957) is also better than that of Model 5.

```python
# Here's something a little more reasonable...
model6_linear_form = 'HP ~ Attack + Speed + Q("Sp. Def") + Q("Sp. Atk")'
# And here we'll add the significant indicators from the previous model
# https://chatgpt.com/share/81ab88df-4f07-49f9-a44a-de0cfd89c67c
model6_linear_form += ' + I(Q("Type 1")=="Normal")'
model6_linear_form += ' + I(Q("Type 1")=="Water")'
model6_linear_form += ' + I(Generation==2)'
model6_linear_form += ' + I(Generation==5)'

model6_spec = smf.ols(formula=model6_linear_form, data=pokeaman_train)
model6_fit = model6_spec.fit()
model6_fit.summary()
yhat_model6 = model6_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model6_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model6)[0,1]**2)
```

**Development of model7_linear_form from model6_linear_form**: Model 7 builds on Model 6 by adding interaction terms between Attack, Speed, Special Defense, and Special Attack. These interactions allow the model to account for more complex relationships, which can improve its predictive performance. The inclusion of these interactions increases the model's complexity. Additionally, the model is adjusted by centering and scaling the continuous variables, which reduces the condition number and improves the model's ability to handle multicollinearity. This adjustment led to improvements in both the in-sample and out-of-sample performance compared to the previous model.

```python
model7_linear_form = 'HP ~ Attack * Speed * Q("Sp. Def") * Q("Sp. Atk")'
model7_linear_form += ' + I(Q("Type 1")=="Normal")'
model7_linear_form += ' + I(Q("Type 1")=="Water")'
model7_linear_form += ' + I(Generation==2)'
model7_linear_form += ' + I(Generation==5)'

model7_spec = smf.ols(formula=model7_linear_form, data=pokeaman_train)
model7_fit = model7_spec.fit()
model7_fit.summary()
yhat_model7 = model7_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model7_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model7)[0,1]**2)
```
